```python
from pyspark.sql.functions import (
    col, when, isnan, isnull, count, mean, stddev, 
    regexp_replace, trim, lower, upper, initcap,
    to_date, to_timestamp, datediff, current_date,
    round as spark_round, lit, coalesce, concat_ws,
    monotonically_increasing_id, row_number, dense_rank
)
```

```python
def analyze_missing_values(df):
    """Analyse complÃ¨te des valeurs manquantes"""
    
    # Compter les nulls par colonne
    missing_counts = df.select([
        count(when(isnull(c) | isnan(c), c)).alias(c) 
        for c in df.columns
    ])
    
    # Calculer les pourcentages
    total_rows = df.count()
    missing_df = missing_counts.toPandas().T
    missing_df.columns = ['missing_count']
    missing_df['missing_percentage'] = (missing_df['missing_count'] / total_rows * 100).round(2)
    missing_df = missing_df[missing_df['missing_count'] > 0].sort_values('missing_count', ascending=False)
    
    print(missing_df)
    
    return missing_df
```
